Este proyecto de iris utilizando machine learning busca resolver ese problema mediante la construcción de un modelo de clasificación que pueda identificar con precisión las tres especies de iris: setosa, versicolor y virginica. Para lograrlo, se utilizará un conjunto de datos de iris etiquetados como

- Id
- SepalLengthCm
- SepalWidthCm 
- PetalLengthCm
- PetalWidthCm
- Species

## 3. Pre procesamiento

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
```

## 4. Entrenamiento ML

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

#Splitting the data for NN
X = df.iloc[:,1:5].values
y = df.iloc[:,5].values
encoder =  LabelEncoder()
y1 = encoder.fit_transform(y)
Y = pd.get_dummies(y1).values

X_train,X_test, Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)

X1 = df.drop(columns=['Species'])
Y1 = df['Species']
x_train, x_test, y_train, y_test = train_test_split(X1, Y1, test_size=0.2)
```

```python
# logistic regression 
from sklearn.linear_model import LogisticRegression
modellog = LogisticRegression()
modellog.fit(X_train, y_train)
print("Accuracy: ",modellog.score(X_test, y_test) * 100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
modelKN = KNeighborsClassifier()
modelKN.fit(X_train, y_train)
print("Accuracy: ",modelKN.score(x_test, y_test) * 100)
```

```python
from sklearn.tree import DecisionTreeClassifier
modelDT = DecisionTreeClassifier()
modelDT.fit(X_train, y_train)
print("Accuracy: ",modelDT.score(X_test, y_test) * 100)
```

## 5. Resultados ML

```python
from sklearn.metrics import confusion_matrix, classification_report

def corr_maz(model):
    
    y_predict = model.predict(x_test).round()
    report = classification_report(y_test, y_predict)
    model_name = type(modelDT).__name__
    score=model.score(x_test, y_test) * 100
    
    labels = ['Setosa', 'Versicolor', 'Virginica']
    
    sns.heatmap(confusion_matrix(y_test, y_predict), linewidth=0.5, annot=True, 
            cmap="YlGnBu", cbar=False, xticklabels=labels, yticklabels=labels)

    plt.ylabel("Etiquetas Reales")
    plt.xlabel("Etiquetas Predecidas")
    plt.title(f"El modelo {model_name} tuvo una precision de {score:.2f}%")
    plt.show()
    
    print("Informe de Clasificación:\n", report)
```

```python
corr_maz(modellog)
```

```python
corr_maz(modelKN)
```

```python
corr_maz(modelDT)
```

## 6. Creación Red Neuronal

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD,Adam


model = Sequential()

model.add(Dense(10,input_shape=(4,),activation='tanh'))
model.add(Dense(8,activation='tanh'))
model.add(Dense(6,activation='tanh'))
model.add(Dense(3,activation='softmax'))

model.compile(Adam(lr=0.04),'categorical_crossentropy',metrics=['accuracy'])

model.summary()
```

```python
model.fit(X_train,y_train,epochs=100)
y_pred = model.predict(X_test)

y_test_class = np.argmax(y_test,axis=1)
y_pred_class = np.argmax(y_pred,axis=1)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test_class,y_pred_class))
print(confusion_matrix(y_test_class,y_pred_class))
```
